# Dataset Descirption

The happiness scores and rankings use data from the Gallup World Poll.

The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. 

The scores are from nationally representative samples for the years 2013-2016 and use the Gallup weights to make the estimates representative. 

A hypothetical country that has values equal to the world’s lowest national averages for each of the six factors. 

They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

# Separating the Features columns and Target Column:

```python
features=new_df.drop("Happiness Score",axis=1)
target=new_df["Happiness Score"]
```

# Regression Test Model 1

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
for i in range(0,500):
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=i)
    lr.fit(features_train, target_train)
    pred_train=lr.predict(features_train)
    pred_test=lr.predict(features_test)
    print(f"At random state {i},the training accuracy is: {r2_score(target_train, pred_train)}")
    print(f"At random state {i},the testing accuracy is: {r2_score(target_test, pred_test)}")
    print("\n")
```

```python
features_train, features_test, target_train, target_test=train_test_split(features, target, test_size=0.25, random_state=178)
```

```python
lr.fit(features_train, target_train)
```

```python
pred_test=lr.predict(features_test)
```

```python
print("The mean absolute error in each measuring data is at : 0.00023557403571844145% -- Which is > 5%. And this is good for our model. ")
mean_absolute_error(target_test, pred_test)
```

```python
print("Here we observe that for this model we get : 99% Accuracy")
r2_score(target_test, pred_test)
```

# Cross-Validation the Model 1


```python
from sklearn.model_selection import cross_val_score
```

```python
lin_score = cross_val_score(lr,features,target,cv=5,scoring='neg_mean_absolute_error')
```

```python
lin_score_n = -lin_score
lin_score_n
```

```python
print("Here we find that the Actual Mean Difference error is 0.0002561304962406936% and not at 0.00023557403571844145% when cross validated at Random State 178 and difference in its accuracy is at 2.05% which is ok in terms of choosing a model ")

np.mean(lin_score_n)
```

```python
0.0002561304962406936-0.00023557403571844145
```
