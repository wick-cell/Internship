This Dataset is from the 2014 Major Baseball league seasons and in this project the main objective is to predict the number of wins for the given Team, based on the 2015 seasons different indicators of success

# Attributes

The factors that determine who will win is as given below:

Runs, At Bats, Hits, Doubles, Triples, Homeruns, Walks, Strikeouts, Stolen Bases, Runs Allowed, Earned Runs, Earned Run Average (ERA), Shutouts, Saves, Complete Games and Errors

-- R - Runs of the Team


-- AB - At Bats performance of te Team whole together


-- H - Hits of a Team


-- 2B - Doubles score rate of a Team


-- 3B - Triples score rate of a Team


-- HR - Home Runs scored by the Team


-- BB - A walk (or base on balls) occurs when a pitcher throws four pitches out of the strike zone


-- SO - The No of Strike outs the team has achieved


-- SB - The No of Stolen bases the team has conquered


-- RA - Runs Allowed is the amount of runs that score against a pitcher


-- ER - An Earned Run is any run that was fully enabled by the offensive team's production in the face of competent play from t         hedefensive team.


-- ERA - The average of earned runs allowed by a pitcher per nine innings


-- CG - The No of completed games by the Team


-- SHO - A Shoutout refers to the act by which a single pitcher pitches a complete game and does not allow the opposing team to         score a run.

-- SV - The No of Saves achieved by the Team


-- E - The No of Errors Committed by the Team


-- W - The No of Wins achieved by the team of that season

# Checking VIF ( Variance Inflation Factor )

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    vif=pd.DataFrame()
    vif["F"] = df.columns
    vif["VIF_Value"] = [variance_inflation_factor(df.values, i)for i in range(df.shape[1])]
    return(vif)

F=df.iloc[:,1:]
F.head()
```

```python
calculate_vif(F)
```

```python
def Plotter(new_df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    new_df[feature].hist()
    plt.subplot(1,2,2)
    Stat.probplot(new_df[feature],dist='norm',plot=pylab)
    plt.show()
```

# Regression Test Model 1

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
for i in range(0,500):
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=i)
    lr.fit(features_train, target_train)
    pred_train=lr.predict(features_train)
    pred_test=lr.predict(features_test)
    print(f"At random state {i},the training accuracy is: {r2_score(target_train, pred_train)}")
    print(f"At random state {i},the testing accuracy is: {r2_score(target_test, pred_test)}")
    print("\n")
```
