```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
driver4 = webdriver.Chrome(r'C:\Users\Rick\OneDrive\Desktop\Softwares\chromedriver.exe')
```

```python
url4 = 'https://www.flipkart.com/'
```

```python
driver4.get(url4)
```

```python
# To automate chrome driver on Exit button

xbutton = driver4.find_element_by_xpath('/html/body/div[2]/div/div/button').click()
```

```python
# To automate chrome driver on Search bar

searchbar = driver4.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
```

```python
# To search for smartphones

searchbar.send_keys("smartphone")
```

```python
# To automate driver on submit button

searchphones = driver4.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button').click()
```

```python
# To find all URL

PhoneURLS = driver4.find_elements_by_xpath('//a[@class="_1fQZEK"]')
```

```python
all_URL = []

for i in PhoneURLS:
     all_URL.append(i.get_attribute("href")) 
```

```python
from tqdm import tqdm
```

```python
Page1phones = []
for i in tqdm(all_URL):
    driver4.get(i)
    Brandandsmartphones = driver4.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span').text
    try:
        Smartphonecolor = driver4.find_element_by_xpath('//div[@class="_2C41yO"]').text
        SmartphoneRAMROM = driver4.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]').text
        SmartphonePrimarySecondaryCamera = driver4.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]').text
        SmartDisplaySize = driver4.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]').text
        SmartPhoneBattery = driver4.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]').text
        SmartPhonePrice = driver4.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[4]/div[1]/div/div[1]').text
        
        Page1phones.append(Smartphonecolor.text)
        Page1phones.append(SmartphoneRAMROM.text)
        Page1phones.append(SmartphonePrimarySecondaryCamera.text)
        Page1phones.append(SmartDisplaySize.text)
        Page1phones.append(SmartPhoneBattery.text)
        Page1phones.append(SmartPhonePrice.text)
    except:
        Page1phones.append('-')
    PhoneData={'Brand':Brandandsmartphones,
           'Color':Smartphonecolor,
           'RAMROM':SmartphoneRAMROM,
           'Camera':SmartphonePrimarySecondaryCamera,
           'Display':SmartDisplaySize,
           'Battery':SmartPhoneBattery,
           'Price':SmartPhonePrice,
            'URL':i}
    Page1phones.append(PhoneData)
```

```python
import pandas as pd

pd.set_option("display.max_colwidth", -1)

pd.DataFrame(Page1phones)
```
